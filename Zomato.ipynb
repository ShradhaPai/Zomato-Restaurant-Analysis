```python
# Ratio of restaurants having table booking vs no table booking
ratio = df["Has Table booking"].value_counts().get("Yes", 0) / df["Has Table booking"].value_counts().get("No", 1)
print("Table booking ratio(Yes/No):", ratio)
```

```python
# Percentage of restaurants providing online delivery
percentage = (df['Has Online delivery'].value_counts().get('Yes', 0) / df['Has Online delivery'].notna().sum()) * 100
print(f"{percentage:.2f}% of restaurants provide online delivery.")
```

```python
# Difference between number of votes between restaurants which delivers and does not deliver
vote_difference = df[df['Has Online delivery'] == 'Yes']['Votes'].sum() - df[df['Has Online delivery'] == 'No']['Votes'].sum()
print(vote_difference)
```

```python
# Top 10 cuisines served across cities
top_10_cuisines = (df['Cuisines'].dropna().str.split(',').explode().str.strip().value_counts().head(10))
print(top_10_cuisines)
```

```python
# Drop missing cuisines
df_clean = df.dropna(subset=['Cuisines'])

# Count number of cuisines per restaurant
df_clean['Cuisine Count'] = df_clean['Cuisines'].str.split(',').apply(len)

# Get max and min
max_cuisines = df_clean['Cuisine Count'].max()
min_cuisines = df_clean['Cuisine Count'].min()

print("Maximum cuisines served by one restaurant:", max_cuisines)
print("Minimum cuisines served by one restaurant:", min_cuisines)
```

```python
df_exploded = df_clean.copy()
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.split(',')
df_exploded = df_exploded.explode('Cuisines')
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.strip()

top_cuisine_per_city = (
    df_exploded.groupby(['City', 'Cuisines'])
    .size()
    .reset_index(name='Count')
    .sort_values(['City', 'Count'], ascending=[True, False])
    .drop_duplicates('City')
)

print(top_cuisine_per_city)
```

```python
bins = [0, 200, 400, 600, 800, 1000, 2000, df['Average Cost for two'].max()]
labels = ['<200', '200-400', '400-600', '600-800', '800-1000', '1000-2000', '2000+']
df['Cost Range'] = pd.cut(df['Average Cost for two'], bins=bins, labels=labels)

cost_distribution = df['Cost Range'].value_counts().sort_index()
print(cost_distribution)
```

```python
# Ratings by Online delivery
df.groupby('Has Online delivery')['Aggregate rating'].mean()
```

```python
# Ratings by Cuisines (Top 10 Cuisines)
df_cuisine = df.dropna(subset=['Cuisines'])
df_cuisine = df_cuisine.copy()
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.split(',')
df_cuisine = df_cuisine.explode('Cuisines')
df_cuisine['Cuisines'] = df_cuisine['Cuisines'].str.strip()
df_cuisine.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
```
